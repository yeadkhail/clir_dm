```python
import pandas as pd
```

```python
df1 = pd.read_json('./datasets/english/daily_star_articles_with_body_20251225_205337.json')
df2 = pd.read_json('./datasets/english/dhakatribune-articles_with_body_20251224_224803.json')
```

```python
# Check the shape and columns of each dataframe
print("df1 shape:", df1.shape)
print("df2 shape:", df2.shape)
print("\ndf1 columns:", df1.columns.tolist())
print("df2 columns:", df2.columns.tolist())
```

```python
# Add source column to identify the origin of each article
df1['source'] = 'daily_star'
df2['source'] = 'dhakatribune'

# Merge both dataframes
merged_df = pd.concat([df1, df2], ignore_index=True)

print(f"Total articles after merging: {len(merged_df)}")
print(f"\nArticles by source:")
print(merged_df['source'].value_counts())
```

```python
# Export merged dataset to CSV
output_filename = './datasets/english/merged_english_articles.csv'
merged_df.to_csv(output_filename, index=False, encoding='utf-8')
print(f"Successfully exported {len(merged_df)} articles to: {output_filename}")
```
