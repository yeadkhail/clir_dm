```python
import pandas as pd
```

```python
df1 = pd.read_json('./datasets/bangla/amardesh-articles_with_body.json')
df2 = pd.read_json('./datasets/bangla/banglatribune_articles_with_body_20251225_213134.json')
df3 = pd.read_json('./datasets/bangla/jugantor-articles_with_body_20251224_232039.json')
```

```python
df1.shape, df2.shape, df3.shape
```

```python
# Check the columns of each dataframe
print("df1 columns:", df1.columns.tolist())
print("df2 columns:", df2.columns.tolist())
print("df3 columns:", df3.columns.tolist())
```

```python
# Add source column to identify the origin of each article
df1['source'] = 'amardesh'
df2['source'] = 'banglatribune'
df3['source'] = 'jugantor'

# Merge all three dataframes
merged_df = pd.concat([df1, df2, df3], ignore_index=True)

print(f"Total articles after merging: {len(merged_df)}")
print(f"\nArticles by source:")
print(merged_df['source'].value_counts())
```

```python
# Display the first few rows and info about the merged dataset
print("Merged dataset info:")
print(merged_df.info())
print("\nFirst few rows:")
merged_df.head()
```

```python
# Export merged dataset to CSV
output_filename = './datasets/bangla/merged_bangla_articles.csv'
merged_df.to_csv(output_filename, index=False, encoding='utf-8')
print(f"Successfully exported {len(merged_df)} articles to: {output_filename}")
```
